```python
#instalando bibliotecas necess√°rias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
!pip install xgboost

```

```python
df = pd.read_csv("store.csv")
```

```python
df.head(10)
```

```python
df.describe()
```

```python


# 1. Crie uma c√≥pia do DataFrame
df_normalizado = df.copy()

# 2. Selecione as colunas
features_para_normalizar = ['Gastos_Publicidade', 'Numero_Funcionarios', 'Numero_Clientes', 'Distancia_Competidor']

# 3. Inicialize o scaler
scaler = MinMaxScaler()

# 4. Aplique o scaler
df_normalizado[features_para_normalizar] = scaler.fit_transform(df[features_para_normalizar])

# 5. Mostre as primeiras 5 linhas do resultado
print("\nDataFrame com features normalizadas:")
display(df_normalizado.head())
```

```python
# Vamos usar o DataFrame 'df_normalizado' que criamos na c√©lula anterior

print("DataFrame ANTES do One-Hot Encoding:")
display(df_normalizado.head())


# 1. Identificar as colunas com texto que precisam ser transformadas
colunas_categoricas = ['Tipo_Loja', 'Sortimento']

# 2. Aplicar o One-Hot Encoding
# O pandas cria as novas colunas e j√° remove as originais ('Tipo_Loja', 'Sortimento')
df_final_processado = pd.get_dummies(df_normalizado, columns=colunas_categoricas, dtype=int)


# 3. Mostrar o resultado final
print("\nDataFrame DEPOIS do One-Hot Encoding (pronto para o modelo):")
display(df_final_processado.head())
```

```python
df_final_processado.head(10)
```

```python
df_final_processado.describe()
```

```python
df_final_processado.info()
```

```python

# 1. Separar as features (X) da vari√°vel alvo (y)
# X = Todas as colunas, exceto a coluna 'Vendas' e a coluna 'Data' original
# y = Apenas a coluna 'Vendas'

X = df_final_processado.drop(['Vendas', 'Data'], axis=1)
y = df_final_processado['Vendas']


# 2. Aplicar a fun√ß√£o train_test_split
# test_size=0.3 define que 30% dos dados ser√£o para teste (e 70% para treino)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=0)


# 3. Verificar o resultado
# para confirmar que a divis√£o ocorreu como esperado.

print("Separa√ß√£o conclu√≠da com sucesso!")
print("-" * 30)
print(f"Total de amostras no dataset: {len(X)}")
print(f"Amostras para treino (70%): {len(X_train)}")
print(f"Amostras para teste (30%): {len(X_test)}")
```

```python
# 1. Inicializar o modelo
modelo_lr = LinearRegression()


X_train_numeric = X_train.select_dtypes(include=np.number)


X_train_numeric = X_train_numeric.fillna(X_train_numeric.median())


# 2. Treinar o modelo com os dados de treino (numeric columns only)
modelo_lr.fit(X_train_numeric, y_train)

```

```python


# 1. Fazer previs√µes nos dados de teste
X_test_numeric = X_test.select_dtypes(include=np.number)

# Lidar com valores ausentes em colunas num√©ricas preenchendo com a mediana dos dados de treinamento
# Usamos a mediana de X_train_numeric para evitar vazamento de dados do conjunto de teste
X_test_numeric = X_test_numeric.fillna(X_train_numeric.median())

# Garanta que as colunas de teste correspondam √†s colunas de treinamento ap√≥s manipular NaNs
# Esta etapa pode ser necess√°ria se a sele√ß√£o de dtypes ou o preenchimento de NaNs resultar em colunas diferentes.
# Por exemplo, se uma coluna num√©rica em test tivesse todos os NaNs, ela poderia n√£o ser inclu√≠da em select_dtypes(include=np.number)
# ou se uma coluna que era num√©rica em train n√£o for num√©rica em test por algum motivo.
# Uma maneira mais segura √© reindexar X_test_numeric para corresponder √†s colunas X_train_numeric.
X_test_numeric = X_test_numeric.reindex(columns=X_train_numeric.columns, fill_value=0)


previsoes_lr = modelo_lr.predict(X_test_numeric)

# 2. Calcular as m√©tricas
mae_lr = mean_absolute_error(y_test, previsoes_lr)
mse_lr = mean_squared_error(y_test, previsoes_lr)
rmse_lr = np.sqrt(mse_lr)
r2_lr = r2_score(y_test, previsoes_lr)

print("--- M√©tricas da Regress√£o Linear ---")
print(f"MAE (Erro M√©dio Absoluto): R$ {mae_lr:,.2f}")
print(f"RMSE (Raiz do Erro Quadr√°tico M√©dio): R$ {rmse_lr:,.2f}")
print(f"R¬≤ (Coeficiente de Determina√ß√£o): {r2_lr:.2%}")
```

```python


# 1. Inicializar o modelo (com random_state para reprodutibilidade)
modelo_dt = DecisionTreeRegressor(random_state=42)

# Selecione apenas colunas num√©ricas de X_train
X_train_numeric = X_train.select_dtypes(include=np.number)

# Lidar com valores ausentes em colunas num√©ricas preenchendo com a mediana
X_train_numeric = X_train_numeric.fillna(X_train_numeric.median())


# 2. Treinar o modelo
modelo_dt.fit(X_train_numeric, y_train)

```

```python



# 1. Fazer previs√µes nos dados de teste

X_test_numeric = X_test.select_dtypes(include=np.number)

# Lidar com valores ausentes em colunas num√©ricas preenchendo com a mediana dos dados de treinamento
# Usamos a mediana de X_train_numeric para evitar vazamento de dados do conjunto de teste
X_test_numeric = X_test_numeric.fillna(X_train_numeric.median())

# Garanta que as colunas de teste correspondam √†s colunas de treinamento ap√≥s manipular NaNs
# Esta etapa pode ser necess√°ria se a sele√ß√£o de dtypes ou o preenchimento de NaNs resultar em colunas diferentes.
# Por exemplo, se uma coluna num√©rica em test tivesse todos os NaNs, ela poderia n√£o ser inclu√≠da em select_dtypes(include=np.number)
# ou se uma coluna que era num√©rica em train n√£o for num√©rica em test por algum motivo.
# Uma maneira mais segura √© reindexar X_test_numeric para corresponder √†s colunas X_train_numeric.
X_test_numeric = X_test_numeric.reindex(columns=X_train_numeric.columns, fill_value=0)


previsoes_dt = modelo_dt.predict(X_test_numeric)


# 2. Calcular as m√©tricas
mae_dt = mean_absolute_error(y_test, previsoes_dt)
mse_dt = mean_squared_error(y_test, previsoes_dt)
rmse_dt = np.sqrt(mse_dt)
r2_dt = r2_score(y_test, previsoes_dt)

print("--- M√©tricas da √Årvore de Decis√£o ---")
print(f"MAE (Erro M√©dio Absoluto): R$ {mae_dt:,.2f}")
print(f"RMSE (Raiz do Erro Quadr√°tico M√©dio): R$ {rmse_dt:,.2f}")
print(f"R¬≤ (Coeficiente de Determina√ß√£o): {r2_dt:.2%}")
```

```python


# 1. Inicializar o modelo (com random_state para reprodutibilidade)
# Use the same random_state for consistency if desired
modelo_xgb = XGBRegressor(random_state=42)

# Selecionar apenas colunas num√©ricas de X_train (j√° feito nas c√©lulas anteriores, mas para maior clareza e robustez)
# X_train_numeric j√° est√° dispon√≠vel nas c√©lulas anteriores, mas vamos garantir que seja usado.
# Se X_train_numeric n√£o estivesse dispon√≠vel, repetir√≠amos os passos aqui.
# Supondo que X_train_numeric e y_train estejam dispon√≠veis e pr√©-processados ‚Äã‚Äã(NaNs manipulados).

# 2. Treinar o modelo
# Use os dados de treinamento num√©ricos pr√©-processados
modelo_xgb.fit(X_train_numeric, y_train)

```

```python


# 1. Fazer previs√µes nos dados de teste
# Selecione apenas colunas num√©ricas de X_test
X_test_numeric = X_test.select_dtypes(include=np.number)

# Lidar com valores ausentes em colunas num√©ricas preenchendo com a mediana dos dados de treinamento
# Usamos a mediana de X_train_numeric para evitar vazamento de dados do conjunto de teste
X_test_numeric = X_test_numeric.fillna(X_train_numeric.median())

# Garanta que as colunas de teste correspondam √†s colunas de treinamento ap√≥s manipular NaNs
# Esta etapa pode ser necess√°ria se a sele√ß√£o de dtypes ou o preenchimento de NaNs resultar em colunas diferentes.
# Por exemplo, se uma coluna num√©rica em test tivesse todos os NaNs, ela poderia n√£o ser inclu√≠da em select_dtypes(include=np.number)
# ou se uma coluna que era num√©rica em train n√£o for num√©rica em test por algum motivo.
# Uma maneira mais segura √© reindexar X_test_numeric para corresponder √†s colunas X_train_numeric.
X_test_numeric = X_test_numeric.reindex(columns=X_train_numeric.columns, fill_value=0)

previsoes_xgb = modelo_xgb.predict(X_test_numeric)

# 2. Calcular as m√©tricas
mae_xgb = mean_absolute_error(y_test, previsoes_xgb)
mse_xgb = mean_squared_error(y_test, previsoes_xgb)
rmse_xgb = np.sqrt(mse_xgb)
r2_xgb = r2_score(y_test, previsoes_xgb)

print("--- M√©tricas do XGBoost ---")
print(f"MAE (Erro M√©dio Absoluto): R$ {mae_xgb:,.2f}")
print(f"RMSE (Raiz do Erro Quadr√°tico M√©dio): R$ {rmse_xgb:,.2f}")
print(f"R¬≤ (Coeficiente de Determina√ß√£o): {r2_xgb:.2%}")
```

```python
# --- An√°lise de Impacto (Regress√£o Linear) ---

# Obter os nomes das features usadas no treino
features_lr = X_train_numeric.columns

# Criar um DataFrame para ver os coeficientes (impacto)
# Usamos o valor absoluto (np.abs) porque a magnitude do n√∫mero √© o que importa,
# n√£o se ele √© positivo ou negativo.
importancia_lr = pd.DataFrame({
    'Feature': features_lr,
    'Importancia (Coeficiente Absoluto)': np.abs(modelo_lr.coef_)
})

# Ordenar pela import√¢ncia
importancia_lr = importancia_lr.sort_values(by='Importancia (Coeficiente Absoluto)', ascending=False)

print("\n--- Impacto das Vari√°veis (Regress√£o Linear) ---")
print(importancia_lr.head(10))
# ---------------------------------------------------
```

```python
# --- An√°lise de Impacto (√Årvore de Decis√£o) ---

# Obter os nomes das features usadas no treino
features_dt = X_train_numeric.columns

# Criar um DataFrame para ver a import√¢ncia
importancia_dt = pd.DataFrame({
    'Feature': features_dt,
    'Importancia': modelo_dt.feature_importances_
})

# Ordenar pela import√¢ncia
importancia_dt = importancia_dt.sort_values(by='Importancia', ascending=False)

print("\n--- Impacto das Vari√°veis (√Årvore de Decis√£o) ---")
print(importancia_dt.head(10))
# ----------------------------------------------------
```

```python
# --- An√°lise de Impacto (XGBoost) ---

# Obter os nomes das features usadas no treino
features_xgb = X_train_numeric.columns

# Criar um DataFrame para ver a import√¢ncia
importancia_xgb = pd.DataFrame({
    'Feature': features_xgb,
    'Importancia': modelo_xgb.feature_importances_
})

# Ordenar pela import√¢ncia
importancia_xgb = importancia_xgb.sort_values(by='Importancia', ascending=False)

print("\n--- Impacto das Vari√°veis (XGBoost) ---")
print(importancia_xgb.head(10))
# -----------------------------------------------
```
